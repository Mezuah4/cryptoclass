# Additional Graphs for Discussion

## Preliminaries

#### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Comparison of Ethereum Model

```python
# Define Dataframe
recs = [{"Model": "Replication", "Accuracy": 0.982, "95%_CI": 0.010},
        {"Model": "Original", "Accuracy": 0.963, "95%_CI": 0.006}]

comp_df = pd.DataFrame.from_dict(recs)
```

```python
comp_df['Accuracy'] = comp_df['Accuracy'] * 100
comp_df['95%_CI'] = comp_df['95%_CI'] * 100
comp_df
```

```python
pl = sns.barplot(x="Model", y="Accuracy", data=comp_df)
pl.errorbar(data=comp_df, x='Model', y='Accuracy', yerr='95%_CI', ls='', lw=2, color="black")
plt.title("Comparison of Ethereum Model Accuracy")
plt.ylabel("Model")
plt.ylabel("Accuracy (%)")
pl.set_ylim(80, 100)
plt.savefig("figures/add/" + "ETH_Comp" + ".pdf", format="pdf", bbox_inches="tight")
plt.show()
```
